15. What is causation? Explain difference between correlation and causation with an example.

Ans- Causation means one event directly results in another, while correlation simply indicates a relationship between two events, where one doesn't necessarily cause the other. For example, while ice cream sales and shark attacks might be correlated (both increase during summer), ice cream sales do not cause shark attacks, but rather both are influenced by the third variable: sunny weather.

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.

Ans- Optimizers are algorithms that update model parameters (like weights and biases) to minimize a loss function during training. Common types include Gradient Descent, Stochastic Gradient Descent (SGD), and adaptive optimizers like Adam and RMSprop, each with varying strengths and weaknesses.
1.Gradient Descent:
Concept:
Iteratively adjusts model parameters in the direction of the steepest descent of the loss function, aiming to find the minimum.
Example:
Imagine a hiker trying to find the lowest point in a valley. Gradient descent is like the hiker taking steps in the direction of the steepest downward slope to reach the bottom.
